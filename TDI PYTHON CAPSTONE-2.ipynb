```python
# Request Channel analysis

# Count requests by medium
counts = df["Submitted via"].value_counts()

# Display counts
print(counts)

# Bar chart
plt.figure(figsize=(10,4))
counts.plot(kind="bar", edgecolor="black")

# Add labels above bars
for i, value in enumerate(counts):
    plt.text(i, value + 0.2, str(value), ha="center", va="bottom")

plt.title("Number of Requests by Medium of Submission")
plt.xlabel("Submitted via")
plt.ylabel("Complaint count")
plt.xticks(rotation=0)
plt.show()
```

```python
#Top 10 states with frequent complaint
# Count complaints by state
top_states = df["State_full"].value_counts().head(10)

# Display the top 10
print(top_states)

# Plot
plt.figure(figsize=(8,4))
ax = top_states.plot(kind="bar", edgecolor="black", color="skyblue")

# Add data labels
for i, value in enumerate(top_states):
    plt.text(i, value + 0.2, str(value), ha="center", va="bottom")

# Formatting
plt.title("Top 10 States with Frequent Complaints")
plt.xlabel("State")
plt.ylabel("Number of Complaints")
plt.xticks(rotation=90)
plt.show()
```

```python
#Timely response analysis
# Count timely responses
counts = df["Timely response?"].value_counts()

# Define colors 
colors = ["#4CAF50", "#F44336","#6C59C2"]   

# Donut chart
plt.figure(figsize=(10,7))
wedges, texts, autotexts = plt.pie(
    counts, 
    labels=counts.index, 
    autopct='%1.1f%%', 
    startangle=90, 
    wedgeprops={'width':0.4},
    colors=colors
)

# Make labels & percentages horizontal
for text in texts + autotexts:
    text.set_rotation(90)   # keep labels flat (horizontal)
    text.set_fontsize(12)

# Formatting
plt.title("Timely Response to Complaints",loc="left")
plt.show()

```

```python
#Year with the highest level of complaint
# Convert "Date Submitted" to datetime
df["Date submitted"] = pd.to_datetime(df["Date submitted"])

# Extract year into a new column
df["Year"] = df["Date submitted"].dt.year

# Count complaints per year
year_counts = df["Year"].value_counts().sort_index()

# Plot line chart
plt.figure(figsize=(8,5))
plt.plot(year_counts.index, year_counts.values, marker="o", linestyle="-", color="skyblue")

# Add data labels
for x, y in zip(year_counts.index, year_counts.values):
    plt.text(x, y + 0.5, str(y), ha="center")

# Formatting
plt.title("Complaints per Year", loc="left")
plt.xlabel("Year")
plt.ylabel("Number of Complaints")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline

# Convert "Date Submitted" to datetime (if not already)
df["Date submitted"] = pd.to_datetime(df["Date submitted"])

# Extract year
df["Year"] = df["Date submitted"].dt.year

# Count complaints per year
year_counts = df["Year"].value_counts().sort_index()

# Create smoother x values
x = year_counts.index.values
y = year_counts.values

# Generate more points for smooth curve
x_smooth = np.linspace(x.min(), x.max(), 300)
y_smooth = make_interp_spline(x, y)(x_smooth)

# Plot smooth line
plt.figure(figsize=(8,5))
plt.plot(x_smooth, y_smooth, color="skyblue")

# Add markers at actual data points
plt.scatter(x, y, color="black")

# Add data labels
for xi, yi in zip(x, y):
    plt.text(xi, yi + 0.5, str(yi), ha="center")

# Formatting
plt.title("Complaints per Year")
plt.xlabel("Year")
plt.ylabel("Number of Complaints")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Step 1: Crosstab
table = pd.crosstab(df['Issue'], df['Timely response?'])

# Step 2: Add total column for sorting
table['Total'] = table.sum(axis=1)

# Step 3: Sort and select top 5
table_sorted = table.sort_values('Total', ascending=False)
top5 = table_sorted.head(5).drop('Total', axis=1)

# Step 4: Plot as grouped bar chart
ax = top5.plot(kind='bar', figsize=(10,10))
plt.title('Top 5 Issues by Timely Response')
plt.xlabel('Issue')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Timely Response')
plt.tight_layout()

# Step 5: Add data labels ABOVE bars
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', padding=3)

    # shift labels above each bar
    for bar in container:
        height = bar.get_height()
        

plt.show()
```

```python
# Total complaints per channel
total = df.groupby("Submitted via")["Timely response?"].count()

# Timely complaints per channel
timely = df[df["Timely response?"] == "Yes"].groupby("Submitted via")["Timely response?"].count()

# Calculate percentage
percentage = (timely / total * 100).sort_values(ascending=False)

# Plot
plt.figure(figsize=(8, 6))
bars = plt.bar(percentage.index, percentage.values, color="lightgreen")

# Add labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, f"{height:.1f}%", 
             ha='center', va='bottom', fontsize=10)

plt.title("Percentage of Timely Responses by Submission Channel")
plt.xlabel("Submission Channel")
plt.ylabel("Timely Response (%)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
